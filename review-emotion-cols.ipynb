```python
import pandas as pd
import numpy as np

books_to_review = pd.read_csv('books_with_emotions.csv')
# List of base emotion names
# emotions = ['anger', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'neutral']

# Select just the columns we want to compare
comparison_df = books_to_review[['title', 'sadness_x', 'sadness_y']].head(10)
comparison_df

```

```python
# 1. Drop all columns ending with _y
y_columns = [col for col in books_to_review.columns if col.endswith('_y')]
books_to_review = books_to_review.drop(columns=y_columns)
books_to_review.head(10)
```

```python
# 2. Rename _x columns to remove the _x suffix
x_columns = [col for col in books_to_review.columns if col.endswith('_x')]
rename_dict = {col: col.replace('_x', '') for col in x_columns}
books_to_review = books_to_review.rename(columns=rename_dict)
books_to_review.head(10)
```

```python
books_to_review.to_csv('books_with_emotions_2.csv', index=False)
```
